```python
import re
import urllib as ub
```

```python
url = 'https://www.listofcountriesoftheworld.com/'
```

```python
dataObj = ub.urlopen(url).readlines()
```

```python
result = re.findall('(?<=<div id="ctry">).*?(?=</div>)', str(dataObj))
```

```python
cleanData = re.findall('(?<=.html">).*?(?=</a>)', str(result))
```
